```python
from google.colab.patches import cv2_imshow
import cv2
import numpy as np
from skimage import segmentation

# Load the medical image
image = cv2.imread('/content/lung ct image.png', cv2.IMREAD_GRAYSCALE)

# Apply binary thresholding
_, binary_image = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)

# Apply Canny edge detection
edges = cv2.Canny(image, threshold1=100, threshold2=200)

# Apply region growing
seed_row, seed_col = 100, 100  # Example seed coordinates
labels = segmentation.flood(image, (seed_row, seed_col), connectivity=2)

# Display the original image, thresholded image, edges, and region growing result
cv2_imshow(image)
cv2_imshow(binary_image)
cv2_imshow(edges)
cv2_imshow(labels)
```

```python
import cv2
import numpy as np
from skimage import segmentation
import matplotlib.pyplot as plt

# Load the medical image
image = cv2.imread('/content/000013 (8).png', cv2.IMREAD_GRAYSCALE)

# Apply binary thresholding
_, binary_image = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)

# Apply Canny edge detection
edges = cv2.Canny(image, threshold1=100, threshold2=200)

# Create a row-wise layout for displaying images
plt.figure(figsize=(10, 5))

# Display the original image
plt.subplot(1, 3, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image - Adenocarcinoma')
plt.axis('off')
plt.show()

# Display the thresholded image
# Create a row-wise layout for displaying images
plt.figure(figsize=(10,5))
plt.subplot(1, 3, 2)
plt.imshow(binary_image, cmap='gray')
plt.title('Thresholded Image')
plt.axis('off')
plt.show()

# Display the edges image
# Create a row-wise layout for displaying images
plt.figure(figsize=(10,5))
plt.subplot(1, 3, 3)
plt.imshow(edges, cmap='gray')
plt.title('Edge Detection')
plt.axis('off')
plt.show()
```
